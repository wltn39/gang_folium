```python
import numpy as np
import pandas as pd
import json
import requests
import branca
```

```python
df = pd.read_csv('서울특별시 강서구 (안심이) CCTV 설치 현황.csv', encoding='cp949')
df.shape
```

```python
from folium.plugins import MarkerCluster

m = folium.Map(
    location=[37.550332499999996,126.84896699999999],
    tiles = 'http://api.vworld.kr/req/wmts/1.0.0/93B6F782-58D4-3D31-AE45-353A7904189D/Base/{z}/{y}/{x}.png',
    attr = 'VWorldBase',
    zoom_start = 12
)

coords = df[['위도','경도']]

marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['위도'], coords['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)
m
```

```python
m.save('강서구_CCTV_basic.html')
```
